```python
import pandas as pd
df = pd.read_csv('http://drive.google.com/uc?export=download&id=1OfqopQz1by_XHGRqK8wMfkOibWf4qUk1')

df.head()
```

```python
from scipy import stats

Pack = df[df['Style'] == 'Pack']
Bowl = df[df['Style'] == 'Bowl']
Tray = df[df['Style'] == 'Tray']
Cup = df[df['Style'] == 'Cup']

f, p = stats.f_oneway(Pack.Stars, Bowl.Stars, Tray.Stars, Cup.Stars)

print('\nF:'+str(f))
print('p:'+str(p))
```
