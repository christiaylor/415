**Return all individuals in the southwest region that have two or more children**

```python
df[(df.region == 'southwest') & (df.children >=2)]
```

**Return all individuals from the northwest region that are either 20 or younger or 60 and older. Sort by age descending.**

```python
df[(df['region'] == 'northwest') & ((df.age <= 20) | (df.age >= 60))].sort_values(by=['age'], ascending = False)
```

**Return all individuals that have charges of 50000 dollars or greater and are from either the northwest, southwest, or southeast region.**

```python
df[((df.region == 'northwest') | (df.region == "southwest") | (df.region == 'southeast')) & (df.charges >= 50000)]
```

```python
regionList = ['northwest', 'southwest', 'southeast']
df[(df.region.isin(regionList)) & (df.charges >= 50000)]
```

**Return all individuals that have charges 50000 or greater and are NOT from either the northwest, southwest, or southeast region.**

```python
df[(~df.region.isin(regionList)) & (df.charges >= 50000)]
```

# **Group By**
